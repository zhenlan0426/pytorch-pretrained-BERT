```python
from bs4 import BeautifulSoup
import pandas as pd
import re
from pytorch_pretrained_bert import BertTokenizer
import pickle
```

```python
train = pd.read_csv("../Data/labeledTrainData.tsv", header=0, \
                    delimiter="\t", quoting=3)
```

```python
test = pd.read_csv("../Data/testData.tsv", header=0, \
                    delimiter="\t", quoting=3)
```

```python
unlabeled = pd.read_csv("../Data/unlabeledTrainData.tsv", header=0, \
                    delimiter="\t", quoting=3)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def Bert_tokenize(example, tokenizer):
    token = tokenizer.tokenize(example)
    #token = ["[CLS]"] + token + ["[SEP]"]
    input_id = tokenizer.convert_tokens_to_ids(token)
    return input_id
```

```python
def precess_review(raw_review):
    review_text = BeautifulSoup(raw_review).get_text()
    review_text = re.sub("[^a-zA-Z]", " ", review_text) 
    return Bert_tokenize(review_text,tokenizer)
```

```python
precessed_reviews = [precess_review(review) for review in train["review"].tolist()]
```

```python
with open('../Data/precessed_reviews.pickle', 'wb') as fp:
    pickle.dump(precessed_reviews, fp)
```

```python
with open('../Data/targets.pickle', 'wb') as fp:
    pickle.dump(train.sentiment.tolist(), fp)
```

```python
precessed_reviews_test = [precess_review(review) for review in test["review"].tolist()]
```

```python
with open('../Data/precessed_reviews_test.pickle', 'wb') as fp:
    pickle.dump(precessed_reviews_test, fp)
```

```python
unlabeled = [precess_review(review) for review in unlabeled["review"].tolist()]
```

```python
with open('../Data/precessed_reviews_unlabeled.pickle', 'wb') as fp:
    pickle.dump(unlabeled, fp)
```
